```python
import pandas as pd
import numpy as np
from functools import reduce

#data movies rotten tomatoes
data = pd.read_csv('rotten_tomatoes_movies.csv', usecols=[1,6,7,8,9,13,15,18])
#rename columns for first dataset
data.columns = ['title', 'genre', 'directors','writers','actors','studio_name','tomato_rating','audience_rating']
#dataset that contains budget
dataBudget = pd.read_csv('budget.csv', usecols=[0,1,6])
dataBudget.columns = ['budget', 'genres' , 'title']
#dataBudget.head(10)

#list of datasets
dfs = [data, dataBudget]
#merging the movies dataset with the budget dataset based on title
df_final = reduce(lambda left,right: pd.merge(left,right,on='title'), dfs)

#data: (title, genre, directors, writers, actors, studio name) labels: (tomoto_rating, audience rating, budget)
df_final = df_final.dropna()  #3920 before dropping null values -> 3830 after dropping those values
#df_final.shape
df_final.head()
```

```python
#iterate through the actors column and split them by the ',' and put them in a list

#fastest way to get length
actorsLen = len(df_final.index) #3

#splitting cells into multiple lists
actorsList = df_final.actors.str.split(', ')

#create a list where actors will be stored
listOfActors = []

#ITERATE THROUGH ONE LIST
#for i in range(len(actorsList[0])):
    #txt = actorsList[0][i].split(', ')
    #print(txt)
   
    #Working --> adding actors
for i in range(len(actorsList)):
    try:
        for j in range(len(actorsList[i])):
            txt = actorsList[i][j].split(', ')
            #print(i)
            listOfActors += txt
            #print(txt)
    except KeyError:
        continue
                
#removing duplicates from listOfActors
setOfActors = list(set(listOfActors))

#len(listOfActors) #-> 141445
#len(setOfActors) #-> 82184
```
